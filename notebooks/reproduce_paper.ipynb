# Reproduce: *The electron as a quantized Kerr resonator* (Knopp 2026)

This notebook regenerates every numerical claim from the paper using the `dinos` package.

Run top-to-bottom; each cell is self-contained.

```python
from dinos import closure, dm, geodesic, cp, casimir, verify, constants as C
```

## §14 — Mass-loop closure (Table 14.2)

```python
result = closure.full_closure()
print(f"m_e       = {result.m_e_MeV:.6f} MeV          (paper: 0.511)")
print(f"σ         = {result.sigma_MeV3:.4e} MeV^3     (paper eq. 65: 2.74e-2)")
print(f"v_bag     = {result.v_bag_MeV:.3f} MeV           (paper eq. 66: 0.43)")
print(f"a         = {result.a_m:.3e} m              (paper eq. 15: 1.9e-13)")
print()
print('Fractional decomposition of m_e c² (Table 14.2):')
f = result.fractions
print(f'  Higgs wall     {f.higgs_wall*100:6.3f}%  (paper 63.880)')
print(f'  Dirac Casimir  {f.dirac_casimir*100:6.3f}%  (paper 35.390)')
print(f'  EM self        {f.em_self*100:6.3f}%  (paper  0.730)')
print(f'  total          {f.total*100:6.3f}%  (paper 100.000)')
```

## §16 — Joint closure: single-point prediction (Remark 16.14)

```python
s = dm.joint_closure()
print(f"λ_H      = {s.lambda_H:.4f}      (paper: 0.129 ± 0.060)")
print(f"v_bag    = {s.v_bag_MeV:.4f} MeV  (paper: ≃0.43)")
print(f"m_*      = {s.m_star_MeV*1000:.1f} keV   (paper: 156 ± 25)")
print(f"y_e      = {s.y_e:.3f}        (paper: ≃1.19)")
print(f"m_* · a  = {s.m_star_times_a:.4f}     (paper: 0.153 ± 0.025)")
print(f"κ = m_*/m_e = {s.kappa:.4f}   (paper eq. 72: 0.602 · λ_H^(1/3))")
```

## §16.6 — Testable predictions P1–P4

```python
pred = dm.predictions()
print('P1  electron-DM scattering σ_eΦ(q→0) =', f"{pred['sigma_eDM_cm2']:.3e} cm²")
print('    (paper quotes ≃ 10⁻³⁴ cm² — the formula as written in eq. 82)')
print()
print('P2  EDGES 21-cm window match        :', pred['edges_match'])
print()
print('P3  ΔF_1^(Φ)(q² = 1 MeV²)           =', f"{pred['dF1_at_q_1MeV']:.3e}")
print()
print('P4  stellar cooling / photon cooling =', f"{pred['stellar_cooling_fraction']:.3e}")
print()
print('Direct detection bounds:')
for k, v in pred['direct_detection'].items():
    print(f'  {k}: {v}')
```

## §7 — Bohr-Sommerfeld ground state (Corollary 7.6)

```python
gs = geodesic.ground_state()
print(f"Ground state (n_φ, n_θ) = ({gs.n_phi}, {gs.n_theta})")
print(f"  → |k| = {gs.k_abs}, j = {gs.j}, m_j = {gs.m_j}")
print('  (paper: 1s_{1/2}, i.e. |k|=1, j=½) — matches hydrogen ground state')
```
